```python
import pandas as pd
```

```python
df = pd.read_csv("mefeatures_100K.csv", dtype={"Name": str})
```

```python
me_columns = [col for col in df.columns if col.endswith('_me')]
morp_columns = [col.replace('_me', '') for col in me_columns]
```

```python
filtered_df = df[df["soma_region"].str.startswith("MB", na=False)]
mefeature_df = filtered_df[me_columns]
morp_df = filtered_df[morp_columns]
morp_df
```

```python
morp_df.to_pickle("morph_data.pkl")
mefeature_df.to_pickle("me_data.pkl")
```
